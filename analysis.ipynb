## MIS Network Optimization Project – Shortest Path Analysis

**Problem:** An e-commerce company operating in Istanbul needs to find the minimum-cost delivery routes from its central warehouse to customer zones, passing through distribution hubs.

**Tool:** NetworkX (Python)

## 5. Managerial Interpretation

The Dijkstra shortest-path algorithm identified the minimum-cost delivery route from the central Istanbul warehouse to each customer zone. Key findings:

- **Hub_Sisli** acts as the most efficient intermediate node, serving two customer zones at low marginal cost.
- **Hub_Kadikoy** connects to the highest-cost customer zone (Pendik), suggesting a potential depot closer to the Asian side could reduce total logistics expenses.
- The total optimized delivery cost across all customer zones is significantly lower than a naive direct-routing approach, demonstrating the value of hub-and-spoke network design in urban e-commerce logistics.
- Decision-makers can use this model to evaluate whether opening a new hub in the Pendik area would yield cost savings that justify the fixed investment.
